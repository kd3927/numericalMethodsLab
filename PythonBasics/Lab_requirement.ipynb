
# Foundation for using Python in a Numerical Methods:

## 1. Basic Python Concepts

### 1.1 Variables and data types

```python
# Integer
a = 10

# Float
b = 3.14

# String
c = "Hello, World!"

# Boolean
d = True

print(a, b, c, d,)
```

### 1.2 Lists and Tuples

```python
# List
my_list = [1, 2, 3, 4, 5]
print(my_list)
print(my_list[0])  # Access first element

# Tuple
my_tuple = (1, 2, 3)
print(my_tuple)
print(my_tuple[1])  # Access second element
```

### 1.3 Dictionaries

```python
# Dictionary
my_dict = {"name": "Numerical Methods", "Lab time(hrs)": 2}
print(my_dict)
print(my_dict["name"])  # Access value by key
```

### 1.4 Functions

```python
# Function definition
def my_function(x):
    return x**2

# Function call
result = my_function(5)
print(result)
```

## 2. Control Flow

### 2.1 Conditional Statements

```python
x = 10
if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")
```

### 2.2 Loops

```python
# For loop
for i in range(5):
    print(i)

# While loop
i = 0
while i < 5:
    print(i)
    i += 1
```

## 4. NumPy Basics

### 4.1 Creating arrays

```python
import numpy as np

# Creating a 1D array
arr1 = np.array([1, 2, 3, 4, 5])

# Creating a 2D array
arr2 = np.array([[1, 2, 3], [4, 5, 6]])

print(arr1)
print(arr2)
```

### 4.2 Array Operations

```python
# Element-wise operations
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

print(arr1 + arr2)  # Addition
print(arr1 * arr2)  # Multiplication

# Matrix multiplication
arr3 = np.array([[1, 2], [3, 4]])
arr4 = np.array([[5, 6], [7, 8]])

print(np.dot(arr3, arr4))  # Dot product
```

### 4.3 Useful NumPy Functions

```python
# Create an array of zeros
zeros = np.zeros((3, 3))

# Create an array of ones
ones = np.ones((2, 2))

# Create an identity matrix
identity = np.eye(3)

# Generate a range of values
range_array = np.arange(0, 10, 2)

print(zeros)
print(ones)
print(identity)
print(range_array)
```

## 5. Plotting with Matplotlib

### 5.1 Basic Plotting

```python
import matplotlib.pyplot as plt

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plotting
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Wave')
plt.show()
```

### 5.2 Subplots

```python
# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Creating subplots
plt.subplot(2, 1, 1)
plt.plot(x, y1)
plt.title('Sine')

plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title('Cosine')

plt.tight_layout()
plt.show()
```
